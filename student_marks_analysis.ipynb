```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {
    'Student': ['Anita', 'Raj', 'Priya', 'Kumar', 'Divya', 'Arjun'],
    'Maths': [85, 72, 90, 65, 88, 76],
    'Science': [78, 80, 92, 70, 85, 68],
    'English': [88, 65, 85, 72, 90, 80],
    'Tamil': [90, 70, 88, 68, 92, 75]
}

df = pd.DataFrame(data)
print(df)
```

```python
df['Total'] = df['Maths'] + df['Science'] + df['English'] + df['Tamil']
df['Average'] = df['Total'] / 4

print(df[['Student', 'Total', 'Average']])
print("\nTop Student:", df.loc[df['Average'].idxmax(), 'Student'])
print("Lowest Scorer:", df.loc[df['Average'].idxmin(), 'Student'])
```

```python
df.plot(x='Student', y=['Maths', 'Science', 'English', 'Tamil'], kind='bar')
plt.title('Student Marks by Subject')
plt.xlabel('Student')
plt.ylabel('Marks')
plt.tight_layout()
plt.show()
```

```python
avg_marks = df[['Maths', 'Science', 'English', 'Tamil']].mean()

avg_marks.plot(kind='pie', autopct='%1.1f%%')
plt.title('Average Marks Distribution by Subject')
plt.tight_layout()
plt.show()
```
